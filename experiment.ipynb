```python
optimizer = torch.optim.AdamW(params=bpe.parameters(), lr=1e-3)
```

```python
for step in range(10000):
    X, Y = dl.load_batch(8, 32)
    logits, loss = bpe(X, Y)

    optimizer.zero_grad(set_to_none=True)

    loss.backward()
    optimizer.step()
```
